```python
# sns.distplot(titanic[titanic['Embarked']=='S']['Age'],hist=False)
# sns.distplot(titanic[titanic['Embarked']=='Q']['Age'],hist=False)
# sns.distplot(titanic[titanic['Embarked']=='C']['Age'],hist=False)
z=titanic['Embarked'].unique()
# print(type(z))


# del z[-1]
z=z[0:3]
colors_arr=["r","b","g"]
print(z.shape[0])

for i in range(0,z.shape[0]):
    sns.distplot(titanic[titanic['Embarked']==z[i]]['Age'],hist=False,color=colors_arr[i])
    print(i)
    

```
