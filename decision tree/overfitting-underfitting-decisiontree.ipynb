```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
def analyser(maxDepth):
    data = pd.read_csv("Social_Network_Ads.csv")
    X=data.iloc[:,2:4]
    y=data.iloc[:,-1].values
    clf = DecisionTreeClassifier(max_depth=maxDepth)
    clf.fit(X,y)

    a=np.arange(start=X.iloc[:,0].min()-1,stop=X.iloc[:,0].max()+1,step=0.1)
    b=np.arange(start=X.iloc[:,1].min()-1,stop=X.iloc[:,1].max()+1,step=100)
    XX,YY = np.meshgrid(a,b)

    input_array=np.array([XX.ravel(),YY.ravel()]).T
    labels = clf.predict(input_array)

    plt.contourf(XX,YY,labels.reshape(XX.shape),alpha = 0.5)
    plt.scatter(X.iloc[:,0],X.iloc[:,1],c=y)
```

```python
analyser(maxDepth=1)
# underfitting
```

```python
analyser(maxDepth=None)
# overfitting
```
