```python
# now using PCA

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=100) 
# n_components -> how many top components we nedded defaiult n dimention -> n pca components 
# NONE  take all components 
```

```python
X_train_trf=pca.fit_transform(X_train)
X_test_trf=pca.transform(X_test)
```

```python
X_train_trf.shape , X_test_trf.shape
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train_trf,y_train)
```

```python
y_pred_trf = knn.predict(X_test_trf)
```

```python
accuracy_score(y_test,y_pred_trf)
```

```python
compontents_numbers = -1
accuracy_percentage = 0
for i in range(1,785):
    pca = PCA(n_components=i)
    X_train_trf=pca.fit_transform(X_train)
    X_test_trf=pca.transform(X_test)
    knn.fit(X_train_trf,y_train)
    y_pred_trf = knn.predict(X_test_trf)
    temp=accuracy_score(y_test,y_pred_trf)
    if temp > 0.95:
        accuracy_percentage=temp
        compontents_numbers=i
        print(temp)
        print(i)
        break;


     
```

```python
print(accuracy_percentage)
print(compontents_numbers)
```
